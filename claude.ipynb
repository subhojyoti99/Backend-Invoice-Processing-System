```python
import aspose.pdf as ap
import os

def pdf_to_image(pdf_path):
    """Convert PDF to image using Aspose.PDF"""
    # Load the PDF document
    document = ap.Document(pdf_path)
    # Create Resolution object
    resolution = ap.devices.Resolution(300)
    # Create JPEG device with specified resolution
    jpg_device = ap.devices.JpegDevice(resolution)
    # Output image path
    image_path = os.path.splitext(pdf_path)[0] + ".jpg"
    # Convert the first page to JPEG
    jpg_device.process(document.pages[1], image_path)
    # Open and return the image
    return image_path
```

```python
import pandas as pd

CSV_FILE = "invoice_data.csv"

def clean_csv_data(file_path):
    try:
        df = pd.read_csv(file_path, encoding="utf-8", sep=",", dtype=str, keep_default_na=False)
        
        # Strip spaces
        df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

        # Remove Euro symbol and fix decimal separators
        df.replace({"€": "", ",": "."}, regex=True, inplace=True)

        return df.to_dict(orient="records")
    
    except Exception as e:
        return {"error": str(e)}

# Example usage:
print(clean_csv_data(CSV_FILE))
```
